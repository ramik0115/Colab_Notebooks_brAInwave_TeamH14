# Preparing X and Y data matrices

```python
## X comprises of the molecular descriptors which are the PubChem fingerprints.

df3_X = pd.read_csv('descriptors_output.csv')
df3_X
```

```python
df3_X = df3_X.drop(columns=['Name'])
df3_X
```

```python
## Y variable.

## Converting IC50 to pIC50.

df3_Y = df3['pIC50']
df3_Y
```

```python
## Combining X and Y variables.


dataset3 = pd.concat([df3_X,df3_Y], axis=1)
dataset3
```
